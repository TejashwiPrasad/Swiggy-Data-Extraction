```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.swiggy.com/city/mumbai/top-rated-collection'
```

```python
r=requests.get(url)
r
```

```python
htmlcontent=r.content
htmlcontent
```

```python
soup=bs(htmlcontent,'html.parser')
soup
```

# TITLE

```python
title=soup.title
print(title.text)
```

# RESTAURANTS NAME

```python
food_name=soup.find_all('div', class_='nA6kb')
len(food_name)
```

```python
FoodName=[]

for i in range(0,len(food_name)): 
    FoodName.append(food_name[i].get_text())
```

```python
FoodName
```

# VARIETY OF FOOD

```python
food_type=soup.find_all('div', class_='_1gURR')
len(food_type)
```

```python
FoodType=[]

for i in range(0,len(food_type)): 
    FoodType.append(food_type[i].get_text())
```

```python
FoodType
```

# RATING OF RESTAURANT

```python
rating=soup.find_all('div', class_='_9uwBC wY0my')
len(rating)
```

```python
Rating=[]

for i in range(0,len(rating)): 
    Rating.append(rating[i].get_text())
```

```python
Rating
```

# FOOD PRICE

```python
price=soup.find_all('div', class_='nVWSi')
len(rating)
```

```python
Price=[]

for i in range(0,len(price)): 
    Price.append(price[i].get_text())
```

```python
Price
```

# DISCOUNT COUPON

```python
DC=soup.find_all('span', class_='sNAfh')
len(DC)
```

```python
discount=[]

for i in range(0,len(DC)): 
    discount.append(DC[i].get_text())
```

```python
discount
```

```python
dataframe={'RESTAURANTS NAME':FoodName,'VARIETY OF FOOD':FoodType,'RATING OF RESTAURANT':Rating,'FOOD PRICE':Price,
           'DISCOUNT COUPON':discount}
```

```python
swiggy=pd.DataFrame(dataframe)
swiggy
```
